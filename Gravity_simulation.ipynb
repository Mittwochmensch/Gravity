```python
import numpy as np
from PIL import Image
import random

Spacevisual = []
timesteps_per_1t = 10000
friction = 1
speed = 0.1
G = 1300 #Gravitational constant

for i in range(1000):
    row = []
    for j in range(1000):
        pixel = []
        for rgb in range(3):
            pixel.append(250)
        row.append(pixel)
    Spacevisual.append(row)
    
class Particle:
    
    def __init__(self, position, mass, velocity, acceleration):
        self.position = position
        self.mass = mass
        self.velocity = velocity
        self.acceleration = acceleration
        
    def update_position(self):
        
        self.position = self.position + (self.velocity * 1 / timesteps_per_1t) 
        
        
    def update_velocity(self):
        
        self.velocity = (self.velocity + (self.acceleration * 1 / timesteps_per_1t)) * friction
        
        
    def update_acceleration(self):
        
        self.acceleration = np.array([0, 0])
        
        for particle in particle_list:
            
            if particle != self:
                
                acceleration = np.array([0, 0])
                acceleration[0] = particle.position[0] - self.position[0]
                acceleration[1] = particle.position[1] - self.position[1]
                acceleration = acceleration * (1 / np.linalg.norm(acceleration**3)) * G * particle.mass
                
                self.acceleration = self.acceleration + acceleration
    
    def test_for_collision(self):
        
        for particle in particle_list:
            
            if particle != self:

                distancevector = self.position - particle.position

                if np.linalg.norm(distancevector) < 3:

                    self.position = (particle.position + self.position)/2
                    self.mass = particle.mass + self.mass
                    self.velocity = (particle.mass * particle.velocity + self.mass * self.velocity)/(particle.mass + self.mass)
                    self.acceleration = np.array([0, 0])

                    particle_list.remove(particle)

```

```python
#define particles
particle_list = []
for i in range(600):
    particle_list.append(Particle(np.array([random.randint(100, 900),random.randint(100, 900)]), random.randint(5, 15), np.array([random.random()*4, random.random()*4]), np.array([0,0])))
```
